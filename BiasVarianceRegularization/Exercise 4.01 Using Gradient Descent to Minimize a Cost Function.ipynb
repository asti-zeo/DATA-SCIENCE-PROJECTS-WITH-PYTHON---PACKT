```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_poly = np.linspace(-3,5,81)
print(x_poly[0:5] , '......' , x_poly[-5:])
```

```python
def cost_fun(x):
    return x*(x-2)
```

```python
y_poly = cost_fun(x_poly)
```

```python
plt.plot(x_poly, y_poly)
plt.xlabel('Parameter val')
plt.ylabel('cost function')
plt.title('Error Surface')
plt.show()
```

```python
def gradient(x):
    return 2*x - 2

x_start = 4.5
learning_rate = 0.25

x_next = x_start - gradient(x_start)*learning_rate
x_next
```

```python
##trial
x_start_2 = 0.75
x_next = -0.75 - gradient(-0.75)*learning_rate
x_next
```

```python
plt.plot(x_poly, y_poly)
plt.plot([x_start, x_next], [cost_fun(x_start), cost_fun(x_next)], marker = 's')
plt.xlabel('Parameter val')
plt.ylabel('cost function')
plt.title('Error Surface')
plt.legend(['cost function', 'error path'])
plt.show()
```

```python
learning_rate = 0.1
iterations = 120
x_path = np.empty(iterations)

x_path[0] = x_start

for iteration_count in range(1,iterations):
    derivative = gradient(x_path[iteration_count -1])
    x_path[iteration_count] = x_path[iteration_count -1] - (derivative*learning_rate)
    
x_path
```

```python
plt.plot(x_poly, y_poly)
plt.plot(x_path, cost_fun(x_path), marker = 's')
plt.xlabel('Parameter val')
plt.ylabel('cost function')
plt.title('Error Surface')
plt.legend(['cost function', 'error path'])
plt.show()
```
